#### Q-9. Given the score of students in multiple exams 

#### Test the hypothesis that the mean scores of all the students are the same. If not, name the student with the highest score. 

```python
import scipy.stats as stats

# Define the data
name = ["Karan", "Deepa", "Karthik", "Chandan", "Jeevan"]
Exam1 = [85, 70, 90, 75, 95]
Exam2 = [90, 80, 85, 70, 92]
Final_exam = [92, 85, 88, 75, 96]

# Perform one-way ANOVA test
f_stat, p_value = stats.f_oneway(Exam1, Exam2, Final_exam)

# Check the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The mean scores of the students are not the same (reject null hypothesis).")
else:
    print("The mean scores of the students are the same (fail to reject null hypothesis).")
```

```python
max_score_index = Final_exam.index(max(Final_exam))
student_with_highest_score = name[max_score_index]
print("The student with the highest score is:", student_with_highest_score)
```
